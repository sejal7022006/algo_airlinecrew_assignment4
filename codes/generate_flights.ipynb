```python
import random, time, sys
import matplotlib.pyplot as plt

# ---------------------------
# FUNCTION: generate_flights
# ---------------------------
def generate_flights(n, start_range=(6,18), duration_range=(1,4)):
    flights = []
    for i in range(n):
        dur = random.randint(duration_range[0], duration_range[1])
        s = random.randint(start_range[0], start_range[1])
        e = s + dur
        flights.append((f"F{i+1}", s, e))
    return flights

# ---------------------------
# TIME & SPACE TESTING
# ---------------------------
sizes = [10, 50, 100, 200, 500]
times = []
spaces = []

for n in sizes:
    start = time.time()
    f = generate_flights(n)
    end = time.time()

    times.append(end - start)
    spaces.append(sys.getsizeof(f))

# ---- TIME GRAPH ----
plt.figure(figsize=(7,4))
plt.plot(sizes, times, marker='o')
plt.title("Generate Flights: Time Complexity")
plt.xlabel("n"); plt.ylabel("Time (sec)")
plt.grid(True); plt.show()

# ---- SPACE GRAPH ----
plt.figure(figsize=(7,4))
plt.plot(sizes, spaces, marker='o', color="red")
plt.title("Generate Flights: Space Complexity")
plt.xlabel("n"); plt.ylabel("Space (bytes)")
plt.grid(True); plt.show()
```
