```python
import time, sys
import matplotlib.pyplot as plt

def flights_overlap(f1, f2, min_rest=0):
    s1,e1 = f1[1], f1[2]
    s2,e2 = f2[1], f2[2]
    if e1 + min_rest <= s2 or e2 + min_rest <= s1:
        return False
    return True

def can_assign(schedule, flight, min_rest=1):
    for f in schedule:
        if flights_overlap(f, flight, min_rest):
            return False
    return True

# ---------------------------
# TIME & SPACE TESTING
# ---------------------------
sizes = [10, 50, 100, 200, 500]
times = []
spaces = []

from random import randint

for n in sizes:
    schedule = [(f"F{i}", randint(6,18), randint(19,25)) for i in range(n)]
    test_flight = ("FT", 10, 12)

    start = time.time()
    can_assign(schedule, test_flight, min_rest=1)
    end = time.time()

    times.append(end - start)
    spaces.append(sys.getsizeof(schedule))

# ---- TIME GRAPH ----
plt.figure(figsize=(7,4))
plt.plot(sizes, times, marker='o')
plt.title("Constraint Checker: Time Complexity")
plt.xlabel("n"); plt.ylabel("Time (sec)")
plt.grid(True); plt.show()

# ---- SPACE GRAPH ----
plt.figure(figsize=(7,4))
plt.plot(sizes, spaces, marker='o', color="purple")
plt.title("Constraint Checker: Space Complexity")
plt.xlabel("n"); plt.ylabel("Space (bytes)")
plt.grid(True); plt.show()
```
