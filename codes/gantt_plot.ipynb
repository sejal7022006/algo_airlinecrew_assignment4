```python
# gantt_plot.py
import matplotlib.pyplot as plt

def plot_gantt(assignments, title="Crew Schedule Gantt"):
    """
    assignments: dict crew -> list of flights (id,start,end)
    """
    crews = list(assignments.keys())
    fig, ax = plt.subplots(figsize=(10, 0.8 + 0.6*len(crews)))
    yticks = []
    ylabels = []
    for i, c in enumerate(crews):
        flights = sorted(assignments[c], key=lambda x: x[1])
        for f in flights:
            start = f[1]; duration = f[2]-f[1]
            ax.barh(i, duration, left=start, height=0.4, align='center')
            ax.text(start + duration/2, i, f[0], va='center', ha='center', color='white', fontsize=8)
        yticks.append(i); ylabels.append(c)
    ax.set_yticks(yticks); ax.set_yticklabels(ylabels)
    ax.set_xlabel("Time"); ax.set_title(title)
    ax.grid(axis='x', linestyle='--', alpha=0.6)
    plt.tight_layout(); plt.show()

# Example usage:
if __name__ == "__main__":
    sample_assign = {
        "C1": [("F1",9,11), ("F4",13,15)],
        "C2": [("F2",10,12)],
        "C3": [("F3",12,13)]
    }
    plot_gantt(sample_assign, title="Sample Gantt")
```
